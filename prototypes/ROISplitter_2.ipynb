```python
import cv2
import cv
import numpy as np
import pylab as pl
import scipy.ndimage
from sleepysnail.preprocessing import ROISplitter
```

```python
grid_img = cv2.imread("grid_test2.jpg")
template = cv2.imread("template.png")
grid_grey = cv2.cvtColor(grid_img, cv.CV_BGR2GRAY)
template = cv2.cvtColor(template, cv.CV_BGR2GRAY)
pl.imshow(grid_img)
```

```python
#pl.imshow(template)
elem = np.ones((11,11),dtype=np.uint8)
grey = cv2.medianBlur(grid_grey,3)
morph = cv2.dilate(grey, elem)
morph = cv2.erode(morph, elem)
morph = cv2.subtract(morph, grey)
morph = cv2.medianBlur(morph, 9)
pl.imshow(morph)
```

```python
def distance_fun(pts, ncols = 4, nrows = 10):
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
    compact_c, _ ,centres_c = cv2.kmeans(pts[:,0],ncols, criteria , 5, flags=cv2.KMEANS_PP_CENTERS)
    compact_r, _ ,centres_r = cv2.kmeans(pts[:,1],nrows, criteria , 5, flags=cv2.KMEANS_PP_CENTERS)
    return log10(compact_c) + log10(compact_r), centres_c.flatten(), centres_r.flatten()

def make_rotation_mat(teta):
    return np.array([
            [cos(teta), -sin(teta)],
            [sin(teta), cos(teta)],
        ])

corners = cv2.cornerHarris(morph, 21, 11, 1e-2,30)
p = 98 # threshold using percentile
decimate = 0.8 # elimitate this proportion sampling = faster
t = np.percentile(corners,p)

rand = numpy.random.uniform(0,1,morph.shape[0] * morph.shape[1]).reshape(morph.shape)
bin_mat = (corners > t) & (rand > decimate)

points = [(j,i) for (i,j), value in numpy.ndenumerate(bin_mat) if value]

best_dist = np.Inf
for teta in np.arange(-np.pi/4, np.pi/4, np.pi/128):    
    rotat_mat = make_rotation_mat(teta)
    rotated_points = dot(points, rotat_mat.T).astype(np.float32)
    dist, centres_c, centres_r = distance_fun(rotated_points)
    if dist < best_dist:
        best_dist = dist
        best_centres_c, best_centres_r = centres_c, centres_r
        invert_teta = -teta 

best_centres_r =  np.sort(best_centres_r)
best_centres_c =  np.sort(best_centres_c)

all_pts = np.array([(c,r) for c, r in itertools.product(best_centres_c,best_centres_r)])
invert_rotat_mat =  make_rotation_mat(invert_teta)
theo_points = dot(all_pts, invert_rotat_mat.T)


toshow = np.copy(grid_img)
test = 10
for i,p in enumerate(theo_points):
    pt = tuple(p.astype(np.int))
    cv2.circle(toshow, pt, 20,(i*10,0,255 - i*10),-1)
    
pl.imshow(toshow)

```
