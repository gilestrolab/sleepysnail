```python
import cv2
import cv
import numpy as np
import pylab as pl
from sleepysnail.preprocessing import ROISplitter
```

```python
grid_img = cv2.imread("grid_test.jpg")
grid_grey = cv2.cvtColor(grid_img, cv.CV_BGR2GRAY)
roi_splitter = ROISplitter(grid_grey)
```

```python
contour_rect_boxes = roi_splitter.rois
tmp = np.copy(grid_img)
cv2.drawContours(tmp,[contour_rect_boxes[0][0]] , -1, (0,255,255),-1)
tmp = tmp/2 + grid_img/2
pl.imshow(tmp)

```

```python
image = grid_grey
print shape(image)

for co, re, bo in [contour_rect_boxes[0]]:
    rot_mat = cv2.getRotationMatrix2D(tuple((np.array(re[0]) - np.array(re[1])) * 2), re[2], 1.0)
    a = np.complex(bo[0][0], bo[0][1])
    b = np.complex(bo[1][0], bo[1][1])
    c = np.complex(bo[2][0], bo[2][1])
    size = (int(np.abs(c-b)), int(np.abs(a-b)))
    
    pts =  np.int0(bo)
    x,y,w,h = cv2.boundingRect(co)
    
    subimg = image[y:y+h,x:x+w]
    cv2.warpAffine(rot_mat, subimg, size)
    
    subimg = cv2.cvtColor(subimg, cv.CV_GRAY2BGR)
    
    pl.imshow(subimg)
    #cv2.rectangle()

#pl.imshow(tmp)    
```
