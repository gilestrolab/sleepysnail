```python
import cv2
import cv
import numpy as np
import pylab as pl
```

```python
grid_img = cv2.imread("grid_test.jpg")
grid_grey = cv2.cvtColor(grid_img, cv.CV_BGR2GRAY)
pl.imshow(grid_img)
```

```python
# here we keep the lines

hline= np.array([[0,0,0], [1,1,1], [0,0,0]],dtype=np.uint8)
vline= np.array([[0,1,0], [0,1,0], [0,1,0]],dtype=np.uint8)


dilated  = cv2.dilate(grid_grey,vline, iterations=10)
vdil  = cv2.erode(dilated,vline, iterations=10)

dilated  = cv2.dilate(grid_grey,hline, iterations=10)
hdil  = cv2.erode(dilated,hline, iterations=10)

dilated  = cv2.dilate(grid_grey,None, iterations=10)

diff_img = cv2.subtract(dilated, hdil)
diff_img = cv2.erode(diff_img, hline, iterations=64)
diff_img = cv2.dilate(diff_img, hline, iterations=64)

diff_img2 = cv2.subtract(dilated, vdil)
diff_img2  = cv2.erode(diff_img2,vline, iterations=64)
diff_img2  = cv2.dilate(diff_img2,vline, iterations=64)

diff_img = cv2.add(diff_img2, diff_img)
diff_img  = cv2.dilate(diff_img,None, iterations=3)
#diff_img_ero  = cv2.erode(diff_img,None, iterations=3)

#diff_img = cv2.subtract(diff_img_dil, diff_img_ero)
#diff_img  = cv2.dilate(diff_img,hline, iterations=10)
#diff_img  = cv2.dilate(diff_img,vline, iterations=10)
#diff_img  = cv2.erode(diff_img,None, iterations=10)

med_img = cv2.medianBlur(diff_img, 101)
diff_img = cv2.subtract(diff_img, med_img)
toshow = cv2.cvtColor(diff_img, cv.CV_GRAY2RGB)

pl.imshow(toshow)
```

```python
_, tmp = cv2.threshold(diff_img, 30, 255,cv.CV_THRESH_BINARY_INV)
toshow = cv2.cvtColor(tmp, cv.CV_GRAY2RGB)
pl.imshow(toshow)
```

```python
_, tmp = cv2.threshold(diff_img, 30, 255,cv.CV_THRESH_BINARY_INV)
contours, hiera = cv2.findContours(tmp, cv.CV_RETR_CCOMP,cv.CV_CHAIN_APPROX_SIMPLE)
#remove holes
contours = [c for (c,h) in zip(contours,hiera[0]) if h[3] == -1 and len(c) >3]
#contours = [c for c in countours if ]
def filter_good_contours(c,min_length=200, max_length=500, min_ar = 3.5, max_ar = 6, max_angle=20):
    rect = cv2.minAreaRect(c)
    box = cv.BoxPoints(rect)
    a = np.complex(box[0][0], box[0][1])
    b = np.complex(box[1][0], box[1][1])
    c = np.complex(box[2][0], box[2][1])
    diag = np.abs(a-c)
    aspect_ratio = np.abs(a-b)/ np.abs(c-b)
    if aspect_ratio < 1:
        aspect_ratio = 1/aspect_ratio
    #print diag
    if diag  > max_length or diag < min_length:
        return False
    
    if aspect_ratio  > max_ar or aspect_ratio < min_ar:
        return False
    if abs(rect[2]% 90) > max_angle and abs(rect[2]) > max_angle:
        return False
    return True

contours = filter(filter_good_contours,contours)
assert(len(contours)==18)
boxes = [np.int0(cv.BoxPoints(cv2.minAreaRect(c))) for c in contours]
rectangles = [cv2.minAreaRect(c) for c in contours]
rot_mats = [cv2.getRotationMatrix2D(tuple((np.array(r[0]) - np.array(r[1])) * 2),
                         r[2], 1.0)  for r in rectangles]

sizes = []
for box in boxes:
    a = np.complex(box[0][0], box[0][1])
    b = np.complex(box[1][0], box[1][1])
    c = np.complex(box[2][0], box[2][1])
    sizes.append((int(np.abs(c-b)), int(np.abs(a-b))))
subimgs = [grid_img[r[0][0]:r[1][0], r[0][1]:r[1][1]] for r in rectangles]
#rotated = [cv2.warpAffine(ai, m, s) for m, s, si in zip(rot_mats, sizes,subimgs)]

tmp = np.copy(grid_img)
cv2.drawContours(tmp, boxes, -1, (0,255,255),-1)
tmp = tmp/2 + grid_img/2
pl.imshow(tmp)
```
