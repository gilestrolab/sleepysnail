```python
import cv2
import cv
import numpy as np
import pylab as pl
from sleepysnail.preprocessing import ROISplitter
```

```python
grid_img = cv2.imread("grid_test.jpg")
grid_grey = cv2.cvtColor(grid_img, cv.CV_BGR2GRAY)
roi_splitter = ROISplitter(grid_grey)
```

```python
tmp = np.copy(grid_img)
cv2.drawContours(tmp, roi_splitter.rois, -1, (0,255,255),-1)
tmp = tmp/2 + grid_img/2
pl.imshow(tmp)
```
